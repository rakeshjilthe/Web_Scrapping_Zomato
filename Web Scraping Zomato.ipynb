### Importing packages

```python
# importing necessary packages
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
```

### Installing Chrome Driver

```python
driver = webdriver.Chrome(executable_path = "C:\\Program Files (x86)\\chromedriver.exe")
```

### Zomato Link of my Location

```python
driver.maximize_window();
driver.get('https://www.zomato.com/pune/order-food-online?delivery_subzone=54554')
```

### Web Scraping

```python
li = driver.find_elements_by_xpath('//a[@class="sc-hdPSEv keaTne"]')
```

```python
li
```

#### Let's check the number of restaurants in the area.

```python
len(li)
```

#### To check all the details of restaurants as link are difficult to view the data.

```python
for value in li:
    print(value.text)
```

### Importing pandas library

```python
import pandas as pd
```

#### Converting all the data to list and from list to DataFrame.

```python
list=[]
for value in li:
    print(value.text)
    list.append(value.text)
```

```python
list
```

```python
df = pd.DataFrame(list)
```

#### Let's check our data now.

```python
df
```

#### Now we got all the data as 'df' using Pandas, but some work need to be done to get it specific.

#### Rename the column as 'All'.

```python
df.columns=["All"]
df
```

#### We can see that every elements is separated by '\n' so, it's better to split using '\n' and giving specific names to columns respectively.

```python
df["All"].str.split('\n',n=1)
```

```python
df[['Name','Rating','Menu', 'Rate', 'Order Placed']]=df['All'].str.split('\n',n=5,expand=True)
df
```

#### Now, we got proper data with separate columns as we expected.
#### But, we can see column with name 'All' is not required, so dropping it.

```python
df. drop('All', axis=1, inplace=True)
```
